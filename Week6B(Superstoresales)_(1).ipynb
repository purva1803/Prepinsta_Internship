 Seasonal Decomposition

```python
import numpy as np
np.random.seed(0)
dates = pd.date_range(start='2020-01-01', periods=365, freq='D')
sales = np.random.normal(loc=0, scale=1, size=365).cumsum() + 1000
df = pd.DataFrame({'Order Date': dates, 'Sales': sales})

def seasonal_decomposition(data, period):
    n = len(data)
    # Compute the seasonal component
    seasonal = np.array([data[i % period] for i in range(n)])
    # Compute the trend component
    trend = np.array([np.mean(data[i:i+period]) for i in range(n)])
    # Compute the residual component
    residual = data - seasonal - trend
    return trend, seasonal, residual
monthly_sales = df.set_index('Order Date').resample('M').sum()
trend, seasonal, residual = seasonal_decomposition(monthly_sales['Sales'].values, 12)

# Plot the decomposition
plt.figure(figsize=(12, 8))

# Original series
plt.subplot(411)
plt.plot(monthly_sales.index, monthly_sales['Sales'], label='Original')
plt.legend(loc='best')

# Trend component
plt.subplot(412)
plt.plot(monthly_sales.index, trend, label='Trend')
plt.legend(loc='best')

# Seasonal component
plt.subplot(413)
plt.plot(monthly_sales.index, seasonal, label='Seasonal')
plt.legend(loc='best')

# Residuals
plt.subplot(414)
plt.plot(monthly_sales.index, residual, label='Residuals')
plt.legend(loc='best')

plt.tight_layout()
plt.show()
```

Sales Variation Over Time

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract month, quarter, and year from 'Order Date'
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter
df['Year'] = df['Order Date'].dt.year

# Group by month and calculate total sales
monthly_sales = df.groupby('Month')['Sales'].sum()

# Group by quarter and calculate total sales
quarterly_sales = df.groupby('Quarter')['Sales'].sum()

# Group by year and calculate total sales
yearly_sales = df.groupby('Year')['Sales'].sum()
fig, axs = plt.subplots(3, 1, figsize=(12, 18))

monthly_sales.plot(ax=axs[0], marker='o', color='b')
axs[0].set_title('Monthly Sales Variation')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Total Sales')

quarterly_sales.plot(ax=axs[1], marker='o', color='g')
axs[1].set_title('Quarterly Sales Variation')
axs[1].set_xlabel('Quarter')
axs[1].set_ylabel('Total Sales')

yearly_sales.plot(ax=axs[2], marker='o', color='r')
axs[2].set_title('Yearly Sales Variation')
axs[2].set_xlabel('Year')
axs[2].set_ylabel('Total Sales')

plt.tight_layout()
plt.show()
```
