Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(car_copy['Make'], bins=20, kde=True, color='skyblue')
plt.xlabel('Make')
plt.ylabel('Ex-Showroom_Price')
plt.title('Distribution of Make')
plt.show()

```

Bar Charts

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Make', data=car_copy)
plt.xticks(rotation=90)
plt.xlabel('Make')
plt.ylabel('Ex-Showroom_Price')
plt.title('Count of Car Makes')
plt.show()

```

Box Plots

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Make', data=car_copy)
plt.xlabel('Make')
plt.title('Box Plot of Make')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Ex-Showroom_Price', data=car_copy)
plt.xlabel('Ex-Showroom Price')
plt.title('Box Plot of Ex-Showroom Price')
plt.show()
```

 Pie Charts


```python
plt.figure(figsize=(8, 8))
make_distribution = car_copy['Make'].value_counts()
plt.pie(make_distribution, labels=make_distribution.index, autopct='%1.1f%%')
plt.title('Distribution of Car Makes')
plt.show()
```

Count Plots


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Fuel_Type', data=car_copy)
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Count of Fuel Types')
plt.show()
```

# **Bivariate Analysis**

Correlation Matrix

```python
plt.figure(figsize=(10, 8))
sns.heatmap(car_copy.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

Scatter Plots


```python
import numpy as np
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Displacement', y='Cylinders', data=car_copy)
plt.title('Scatter Plot of Sales vs. Price')
plt.show()
```

Pair Plots


```python
plt.figure(figsize=(10, 8))
sns.pairplot(car_copy, hue='Fuel_Type')
plt.title('Pair Plot ')
plt.title('Pair Plot of Numeric Variables in Car Dataset')
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()
```

Bar Chart

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='Model', y='Cylinders', hue='Fuel_Type', data=car_copy)
plt.xlabel('Model')
plt.ylabel('Cylinders')
plt.title('Bar Chart of Cylinders by Model with Fuel Type Hue')
plt.xticks(rotation=90)
plt.legend(title='Fuel Type')
plt.show()
```

 Heatmaps


```python
plt.figure(figsize=(10, 8))
sns.heatmap(car_copy.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

Joint Plots


```python
sns.jointplot(x='Cylinders', y='Doors', data=car_copy, kind='scatter')
plt.suptitle('Joint Plot ', y=1.02)
plt.show()
```

 Box Plots

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Make', y='Model', hue='Fuel_Type', data=car_copy, palette='Set3', linewidth=2)
plt.xlabel('Make', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.title('Box Plot', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Fuel_Type', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Categorical Plots


```python
plt.figure(figsize=(10, 6))
sns.catplot(x='Fuel_Type', y='Make', data=car_copy, kind='bar')
plt.xlabel('Fuel Type')
plt.ylabel('Make')
plt.title('Categorical Plot of Price by Fuel Type')
plt.xticks(rotation=90)
plt.show()
```

Violin Plots


```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Ex-Showroom_Price', y='Cylinders', data=car_copy)
plt.xlabel('Ex-showroom price')
plt.ylabel('cylinder')
plt.title('Violin Plot of Price by Fuel Type')
plt.show()
```

Stacked Bar Charts


```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Make', y='Model',  data=car_copy, estimator=sum)
plt.xlabel('Make')
plt.ylabel('Model')
plt.title('Stacked Bar Chart')
plt.xticks(rotation=90)
plt.show()
```
